1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?
jawab:
*Tanpa Matplotlib:

1. Tanpa menggunakan Matplotlib, kita hanya dapat melihat gambar seperti melihat foto di kertas atau layar komputer.
2. Gambar akan muncul seperti biasanya, dan kita hanya bisa melihatnya tanpa bisa melakukan apa-apa. Kita tidak bisa mengubah ukurannya atau menambahkan hal-hal baru ke dalam gambar.

*Dengan Matplotlib:

1. Dengan Matplotlib, kita dapat mengubah cara gambar itu muncul di layar komputer.
2. Kita bisa memilih seberapa besar gambar itu, seperti membesarkannya atau mengecilkannya. Kita juga bisa memberi judul pada gambar tersebut, seperti memberi nama foto kita.
3. kita bisa menambahkan gambar-gambar kecil lainnya atau tulisan-tulisan di atas gambar itu. Kita bisa membuat gambar kita sendiri di atas gambar asli.
Jadi, Matplotlib adalah alat yang membantu kita mengontrol dan bermain dengan gambar-gambar di komputer kita. Kita bisa membuat gambar lebih besar, memberi nama pada gambar, atau bahkan menggambar sesuatu di atas gambar itu. Semua ini tidak bisa kita lakukan jika kita hanya melihat gambar biasa di layar komputer atau kertas.



2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?
* perbedaan utamanya adalah ukuran atau ruang yang kita miliki untuk membuat gambar hitam. int32 memberi kita lebih banyak ruang daripada int16.
* pengaruhnya adalah Ketika kita membuat gambar hitam dengan int16, kita mungkin terbatas dalam ukuran gambar yang dapat kita buat. Namun, dengan int32, kita memiliki lebih banyak ruang untuk membuat gambar hitam yang lebih besar dan lebih besar lagi.
